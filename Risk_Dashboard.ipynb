```python
import pandas as pd
import panel as pn

pn.extension()

# Exemple de donnes (remplacez-le par votre propre DataFrame)
# ALL_Portfolio = pd.read_csv("path_to_your_data.csv")

# Crer le widget de slection du portefeuille
Portfolio_code = list(ALL_Portfolio['Code portefeuille'].unique())
dropdown = pn.widgets.Select(name='Slectionner un portefeuille', options=Portfolio_code)

# Crer un panneau pour afficher le DataFrame
df_pane = pn.pane.DataFrame(pd.DataFrame())  # Initialiser un DataFrame vide

df_Total_pane = pn.pane.DataFrame(pd.DataFrame())  # Initialiser un DataFrame vide

# Fonction pour mettre  jour le portefeuille slectionn  partir du dropdown
def update_selected_portfolio(event):
    portfolio_code = event.new  # Valeur slectionne dans le dropdown
    Selected_Portfolio = ALL_Portfolio[
        ALL_Portfolio['Code portefeuille'] == portfolio_code
    ].copy()

    # Suppression des colonnes indsirables
    columns_to_remove = [
        'Dpositaire', 'Code gestionnaire', 'Mode de gestion', 'Profil de gestion', 
        'Taux', 'Echance', 'Intrts courus', 'Intrts courus en  (mon. rf.)', 
        'PP non ral. en % Orig.', 'PP non ral. en %  (mon. rf.)'
    ]
    Selected_Portfolio.drop(columns=columns_to_remove, inplace=True, errors='ignore')

    # Calculer la valeur totale du march du portefeuille slectionn
    total_selected_portfolio = Selected_Portfolio['Valeur march en  (mon. rf.)'].sum()

    # Ajouter la colonne 'Pondration'
    Selected_Portfolio['Pondration'] = (
        Selected_Portfolio['Valeur march en  (mon. rf.)'] / total_selected_portfolio
    ).round(2)

    df_Total_pane.object = Selected_Portfolio
    

    # Exclusion des liquidits
    Selected_Portfolio_Ex_Cash = Selected_Portfolio[Selected_Portfolio['Catgorie'] != 'Liquidits'].copy()

    # Recalculer la valeur totale de march aprs exclusion des liquidits
    total_selected_portfolio_Ex_Cash = Selected_Portfolio_Ex_Cash['Valeur march en  (mon. rf.)'].sum()

    # Recalculer les pondrations sans les liquidits
    Selected_Portfolio_Ex_Cash['Pondration'] = (
        Selected_Portfolio_Ex_Cash['Valeur march en  (mon. rf.)'] / total_selected_portfolio_Ex_Cash
    ).round(2)

    # Mettre  jour l'affichage du DataFrame avec le tableau actualis
    df_pane.object = Selected_Portfolio_Ex_Cash  # Mettre  jour l'objet avec les donnes actualises

# Surveiller les changements dans le menu droulant
dropdown.param.watch(update_selected_portfolio, 'value')

# Afficher le menu droulant et le DataFrame

test_2 = pn.Column(dropdown, df_Total_pane)

test_2
```

```python
# Filtrer les prix de clture pour le portefeuille slectionn
Closing_price_selected_portfolio = ALL_CLOSING_Price[
    ALL_CLOSING_Price['Code'].isin(df_pane.object['Code'])
]

# Slectionner les X derniers jours (Historique_Vol)
Histo_Closing_price_selected_portfolio = Closing_price_selected_portfolio.iloc[:, -Historique_Vol:]

# Remplir les valeurs manquantes (forward fill)
Histo_Closing_price_selected_portfolio = Histo_Closing_price_selected_portfolio.ffill()

# Calculer les variations journalires en pourcentage
Histo_Closing_price_selected_portfolio_percent = (
        Histo_Closing_price_selected_portfolio.pct_change(axis=1, fill_method=None) * 100
).round(2)

# Ajouter la colonne 'Code' pour garder une rfrence des actifs
Histo_Vol_ISIN_percent = pd.concat(
    [Closing_price_selected_portfolio['Code'], Histo_Closing_price_selected_portfolio_percent],
    axis=1
)

# Calculer la volatilit (cart-type des variations) pour chaque actif
Histo_Vol_ISIN_percent['Volatilit'] = Histo_Vol_ISIN_percent.iloc[:, 1:].std(axis=1).round(2)



```

```python
# Transposer les variations pour aligner les actifs en colonnes
variations_transposes = Histo_Vol_ISIN_percent.iloc[:, 1:-1].T  # Exclure 'Code' et 'Volatilit'

# Nommer les colonnes de la matrice transpose avec les ISIN (valeurs de la premire colonne)
variations_transposes.columns = Histo_Vol_ISIN_percent['Code']

# Calculer la matrice de corrlation entre les actifs
matrice_corr = variations_transposes.corr()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crer une figure et un axe pour le graphique
Correlation_chart, ax = plt.subplots(figsize=(10, 8))

# Gnrer la heatmap
sns.heatmap(
    matrice_corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    xticklabels=matrice_corr.columns,
    yticklabels=matrice_corr.index,
    ax=ax
)

# Ajouter un titre
ax.set_title("Matrice de Corrlation des Actifs (ISIN)")
```

```python
import numpy as np

# Assurez-vous que 'Pondration' existe dans df_pane.object
Ponderation_Ex_Cash = df_pane.object['Pondration'] / 100  # Diviser les pondrations en pourcentage par 100

# Supprimer la colonne "Code" (identifiant)
Price_Variations = Histo_Vol_ISIN_percent.drop(columns=['Code'])

# Transposer la matrice
Price_Variations_T = Price_Variations.copy().T

# Calculer la matrice de covariance annualise
Matrix_Cov = Price_Variations_T.cov() * 252

# Calcul de la variance du portefeuille
Variance_Portfolio_Ex_Cash = np.dot(
    Ponderation_Ex_Cash, np.dot(Matrix_Cov.values, Ponderation_Ex_Cash)
)

# Calculer la volatilit hors liquidits
Volatilite_Selected_Portfolio_Ex_Cash = np.sqrt(Variance_Portfolio_Ex_Cash).round(4)

# Filtrer les lignes correspondant aux liquidits
Cash = df_pane.object[df_pane.object['Catgorie'] == 'Liquidits']

# Calculer la pondration totale des liquidits
Ponderation_Cash = Cash['Pondration'].sum()

# Calculer la volatilit totale du portefeuille
Volatilite_Selected_Portfolio = (1 - Ponderation_Cash) * Volatilite_Selected_Portfolio_Ex_Cash

Volatilite_Selected_Portfolio


```
