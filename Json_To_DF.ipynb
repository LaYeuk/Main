```python
import pandas as pd

# Chemin vers le fichier JSON
chemin_fichier = r"P:\Concept\2024_ALL"



# Affichage des données pour vérification
```

```python
import pandas as pd
import json

# Charger le fichier JSON en DataFrame en gérant l'encodage
def load_json_to_dataframe(json_file, encoding="utf-8"):
    try:
        with open(json_file, "r", encoding=encoding) as file:
            data = json.load(file)  # Charger le JSON en dictionnaire

        if "InvStatistic" in data:
            records = data["InvStatistic"]

            all_keys = set()
            for record in records:
                all_keys.update(record.keys())

            df = pd.DataFrame(records, columns=list(all_keys))  # Convertir en DataFrame avec toutes les clés comme colonnes
        else:
            df = pd.DataFrame(data)  # Fallback si '.InvStatistic' n'existe pas

        print("JSON chargé avec succès !")
        return df
    except UnicodeDecodeError:
        print(f"Erreur d'encodage. Essayez avec un autre encodage, par exemple 'ISO-8859-1'.")
        return None
    except Exception as e:
        print(f"Erreur lors du chargement du JSON : {e}")
        return None

# Exemple d'utilisation
json_file = chemin_fichier
df = load_json_to_dataframe(json_file, encoding="latin1")
```

```python
grouped_df = df.groupby(['InCharge','PeriodName']).sum(numeric_only=True).reset_index()

# Afficher les résultats
grouped_df
```

```python
import pandas as pd

# Assuming grouped_df is already defined
summary_df = grouped_df[['InCharge', 'InvTotalAmtRef', 'AssetAmount']].copy()

# Convert 'AssetAmount' to numeric, coercing errors to NaN if any
summary_df['AssetAmount'] = pd.to_numeric(summary_df['AssetAmount'], errors='coerce')

# Calculate the sum of BOTH InvTotalAmtRef and AssetAmount (handle potential NaNs)
total_inv = summary_df['InvTotalAmtRef'].sum()
total_asset = summary_df['AssetAmount'].sum()

# Create the total row dictionary with separate entries for each column
total_row = {'InCharge': 'Total', 'InvTotalAmtRef': total_inv, 'AssetAmount': total_asset}

summary_df = pd.concat([summary_df, pd.DataFrame([total_row])], ignore_index=True)

# Format 'AssetAmount' column to standard numeric format (no scientific notation)
summary_df['AssetAmount'] = summary_df['AssetAmount'].apply(lambda x: f'{x:.0f}' if pd.notna(x) else x)

print(summary_df)
```
